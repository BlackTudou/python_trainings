| 绘图类型 | 函数 | 主要特点 / 用途 |
|----------|------|----------------|
| 关系图（Relational Plot） | `sns.relplot()` | 比 `plt.plot` 更强大，可自动计算两组数据关系，支持分组、色彩映射等 |
| 分类图（Categorical Plot） | `sns.catplot()` | 观察不同类别的数据分布情况 |
| 分类分布图（Distribution per Category） | `sns.boxplot()` | 先看整体分类，再看每个分类下数据分布情况 |
| 分类统计图（Count Plot） | `sns.countplot()` | 统计每个类别的数量，直观了解类别分布 |
| 线性回归图（Regression Plot） | `sns.regplot()` | 绘制散点并拟合线性回归，直观查看变量间线性关a系 |
| 热力图（Heatmap） | `sns.heatmap()` | 可视化矩阵或特征间相关性，用颜色深浅表示数值大小或相关程度 |

## 绘制关系图

```python
import seaborn as sns
# ----------------------
# 加载 Seaborn 自带的 fmri 数据集
# ----------------------
fmri = sns.load_dataset("fmri")  # 返回一个 pandas DataFrame，包含功能性磁共振成像（fMRI）实验数据

# 查看数据集的前几行
fmri.head()  # 显示 DataFrame 的前 5 行，方便快速了解数据结构和内容

# 数据集包含功能性磁共振成像(fMRI)实验结果，每行表示一次测量
# -----------------------------------------
# subject   : 受试者编号，例如 's13', 's5'
# timepoint : 测量时间点，整数，例如 18, 14
# event     : 实验事件类型，例如 'stim' 表示刺激
# region    : 脑区，例如 'parietal'（顶叶）
# signal    : 测量到的信号值，浮点数，例如 -0.017552, -0.080883
```

```python
import seaborn as sns

# ----------------------
# 使用 Seaborn 绘制折线图（关系图）
# ----------------------
sns.relplot(
    x="timepoint",   # 横坐标：时间点
    y="signal",      # 纵坐标：脑信号值
    kind="line",     # 图形类型为折线图
    hue="event",     # 不同事件类型使用不同颜色区分
    col="event",     # 根据事件类型生成多个子图（每个事件一列）
    data=fmri        # 数据来源：fmri 数据集
)
```

## 绘制分类图

```python
import seaborn as sns

# ----------------------
# 绘制分类图（Categorical Plot）
# ----------------------
sns.catplot(
    x="pclass",    # 横坐标：船舱等级（1/2/3），属于分类变量
    y="age",       # 纵坐标：乘客年龄，数值型变量
    data=titanic   # 数据来源：titanic 数据集
)
```

## 绘制箱型图

```python
import seaborn as sns

# ----------------------
# 绘制箱型图（Box Plot）
# ----------------------
sns.boxplot(
    x="pclass",   # 横坐标：船舱等级（分类变量 1/2/3）
    y="age",      # 纵坐标：乘客年龄（数值变量）
    data=titanic  # 数据来源：titanic 数据集
)
# 显示数据的中位数、上下四分位数（Q1、Q3）、异常值
```

## 绘制条形图

```python
import seaborn as sns

# ----------------------
# 绘制条形图（Bar Plot）
# ----------------------
sns.barplot(
    x='pclass',    # 横坐标：船舱等级（1/2/3），属于分类变量
    y='survived',  # 纵坐标：幸存比例（数值型变量）
    data=titanic   # 数据来源：titanic 数据集
)
# 柱子的高度表示该舱位的乘客平均幸存率
```

## 绘制热力图

机器学习中使用最多

```python
# 热力图
flights = sns.load_dataset("flights")
flights.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# ----------------------
# 将长格式数据转换为宽格式数据
# ----------------------
# flights 数据集包含 year, month, passengers 列（每行一个月）
# pivot 将数据透视为矩阵形式：
#   - index: 行索引，这里用月份
#   - columns: 列索引，这里用年份
#   - values: 填充单元格的值，这里是乘客数量
f = flights.pivot(index="month", columns="year", values="passengers")

# ----------------------
# 绘制热力图
# ----------------------
plt.figure(figsize=(10, 8))  # 设置图表大小，宽10英寸，高8英寸

sns.heatmap(
    f,              # 输入矩阵数据
    annot=True,     # 在每个单元格中显示数值
    fmt="d",        # 数值格式化为整数
    cmap="YlGnBu"   # 配色方案，从黄色到绿色再到蓝色 其它颜色： YlGnBu、coolwarm、RdBu
)

# 设置图表标题和轴标签
plt.title("Flights Passenger Count by Year and Month")  # 标题
plt.xlabel("Year")  # x轴标签
plt.ylabel("Month") # y轴标签

# 显示图表
plt.show()
```

## 绘制相关系数矩阵的热力图

```python
import seaborn as sns
import matplotlib.pyplot as plt

# -----------------------------
# 绘制相关系数矩阵的热力图
# -----------------------------

plt.figure(figsize=(10, 10))  # 设置整个图表的大小（宽10，高10英寸）

# f.corr() 计算 DataFrame f 中各列之间的相关系数矩阵
# 相关系数范围 [-1, 1]：
#   1  → 完全正相关
#   0  → 无相关
#  -1 → 完全负相关

# sns.heatmap() 绘制热力图
# 参数说明：
#   annot=True  → 在每个格子中显示具体数值
#   cmap='coolwarm' → 使用蓝红渐变配色，红色表示正相关，蓝色表示负相关
sns.heatmap(f.corr(), annot=True, cmap='YlGnBu') # cmap='YlGnBu'

# 显示图表
plt.show()
```
