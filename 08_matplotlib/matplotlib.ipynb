## plot函数参数

| 参数       | 全称          | 作用       | 常见取值                                            |
| -------- | ----------- | -------- | ----------------------------------------------- |
| `ls`     | `linestyle` | 线型       | `'-'` 实线、`':'` 点虚线、`'--'` 破折线、`'-.'` 点划线        |
| `lw`     | `linewidth` | 线宽       | 数值，如 `1.5`                                      |
| `label`  | 标签          | 图例显示的文字  | `"sin(x)"`                                      |
| `color`  | 颜色          | 设置线条颜色   | `'r'` 红、`'g'` 绿、`'b'` 蓝、`'y'` 黄、`'k'` 黑、`'w'` 白 |
| `marker` | 标记点         | 每个数据点的样式 | `"."` 点、`","` 像素点、`"o"` 圆点、`"v"` 下三角、`"*"` 星号   |


plt.plot(x, y, 'r--')                # 简写形式：红色虚线

```python
# 导入 numpy 库并简写为 np
# 这里主要用来生成等差数列
import numpy as np  

# 新建一个画布（figure）
# figsize=(8,3) 设置画布大小，宽 8 英寸，高 3 英寸
# dpi=100 表示分辨率，每英寸 100 像素
plt.figure(figsize=(8, 3), dpi=100)  

# 生成等差数列作为横坐标 x
# np.linspace(start, stop, num=50)
# 默认 num=50，表示在 [0, 2π] 区间内生成 50 个等间距的点
x = np.linspace(0, 2 * np.pi)  

# 根据横坐标计算正弦函数值
y = np.sin(x)  

# 绘制曲线
# x, y 为数据
# color="g" 表示绿色线条（green） k代表黑色、-.代表点划线
plt.plot(x, y, color="k")  

# # 打开网格线，便于观察曲线走势
plt.grid()

plt.show()
```

```python
# 绘制多条线，设置画布背景
import numpy as np
# 新建一个画布（figure）
# figsize=(6,3)   → 画布宽 6 英寸，高 3 英寸
# dpi=100        → 分辨率，每英寸 100 像素
plt.figure(figsize=(6, 3), dpi=100)
# 下面两行是可选写法，可以设置画布背景颜色
# plt.figure(figsize=(6,3), dpi=100, facecolor="r")
# plt.figure(figsize=(6,3), dpi=100, facecolor="red")

# 生成等差数列作为横坐标，范围 [0, 2π]，默认 50 个点
x = np.linspace(0, 2 * np.pi)

# 定义两组函数
y = np.sin(x)   # 正弦曲线
z = np.cos(x)   # 余弦曲线

# 绘制第一条曲线
# x, y → 数据
# color="b" → 蓝色线条（blue）
plt.plot(x, y, color="b",label="sinx")

# plt.show() #  → 立即显示图形

# 绘制第二条曲线
# 方式 2：简写形式 → 'r:' 表示红色点虚线
plt.plot(x, z, "r:",label = "cosx")  # 简写形式 → 'r:' 表示红色点虚线
# plt.plot(x,z,color="r",ls=":")
plt.legend()
```

```python
# 多图布局
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
x = np.linspace(0, 10, 100) # 生成横坐标 x，从0到10，100个等间距点
y1 = np.sin(x)
y2 = np.cos(x)

# 创建一个2x2的子图布局
plt.figure(figsize=(10, 6))  # 设置整个图的大小

# 第一个子图
plt.subplot(2, 2, 1)  # 第一个子图位于2x2布局的第1个位置，第三个参数
plt.plot(x, y1, 'r-')  # 绘制sin曲线
plt.title('Sin Function')  # 设置标题

# 第二个子图
plt.subplot(2, 2, 2)  # 第二个子图位于2x2布局的第2个位置
plt.plot(x, y2, 'g-')  # 绘制cos曲线
plt.title('Cos Function')  # 设置标题

# 第三个子图
plt.subplot(2, 2, 3)  # 第三个子图位于2x2布局的第3个位置
plt.plot(x, y1 + y2, 'b-')  # 绘制sin+cos曲线
plt.title('Sin + Cos Function')  # 设置标题

# 第四个子图
plt.subplot(2, 2, 4)  # 第四个子图位于2x2布局的第4个位置
plt.plot(x, y1 - y2, 'm-')  # 绘制sin-cos曲线
plt.title('Sin - Cos Function')  # 设置标题

# 调整子图之间的间距
plt.tight_layout() # 自动调整子图参数，使子图标题、坐标轴标签不重叠

# 显示图形
plt.show()
```

```python
# 图形嵌套
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
x = np.linspace(0, 10, 100)
y = np.sin(x)
# 创建外层图形
fig = plt.figure(figsize=(8, 6))

# -----------------------------
# 绘制外层主图
# -----------------------------
plt.plot(x, y, 'y-')              # 黄色实线绘制 sin 曲线
plt.title('Main Plot')            # 设置主图标题

# -----------------------------
# 创建嵌套子图（Inset）
# -----------------------------
# 设置嵌套子图在画布中的位置
# [left, bottom, width, height]，取值均为 0~1，表示比例
left, bottom, width, height = [0.4, 0.4, 0.25, 0.25]

# 在外层画布中添加新的坐标轴（axes）
axes = fig.add_axes([left, bottom, width, height])

# 在嵌套子图中绘制曲线
axes.plot(x, y, 'r-', label='Sin Function (Inset)')  # 红色实线
# axes.set_title('Inset Plot')                        # 设置嵌套子图标题
plt.title('Inset Plot')

# 显示图例
plt.legend()
# 显示图形
plt.show()
```

```python
# 生成数据便于后续绘图
import pandas as pd           # 用于处理表格数据
import numpy as np            # 用于数值计算
import matplotlib.pyplot as plt  # 用于绘图

# 创建基础数据
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']  # 12个月
students = ['张三']                                   # 学生名单
subjects = ['语文', '数学', '英语']                   # 科目列表

# 生成递增成绩数据
# np.linspace(start, stop, num) 用于生成等间距数列
# 这里每门科目成绩从 60 增加到 95，共 12 个点
data = np.column_stack([ # 把三门科目的数据 按列合并，生成一个 12×3 的数组，方便后续转换成 DataFrame
    np.linspace(60, 95, 12).astype(int),  # 语文成绩递增  使用astype(int) 将成绩取整，保证都是整数
    np.linspace(65, 98, 12).astype(int),  # 数学成绩递增
    np.linspace(62, 97, 12).astype(int)   # 英语成绩递增
])

# 将数据转换为 DataFrame
df = pd.DataFrame(data, columns=subjects, index=months)

# 保存数据到 Excel 文件
excel_file = '张三成绩.xlsx'
df.to_excel(excel_file)
print("数据已保存到 Excel 文件:", excel_file)

# 查看生成的数据
df.head()
```

```python
# 绘制折线图
plt.figure(figsize=(10, 6))

x = df.index # 获取月份作为 x 轴数据
y1 = df['语文'] # 获取张三在语文科目上的成绩作为 y 轴数据
y2 = df['数学']
y3 = df['英语']

plt.plot(x, y1, marker='o', color="r",ls='-',label="语文")
plt.plot(x, y2, marker='*', color="b",ls='-',label="数学")
plt.plot(x, y3, marker='v', color="k",ls='-',label="英语")

plt.title('张三成绩月变化情况')
plt.xlabel('月份')
plt.ylabel('分数')
plt.legend() # 添加图例
plt.grid(True) # 网格线

# 显示图形
plt.xticks(rotation=45) # 将 X 轴刻度标签的角度旋转为 45 度
plt.tight_layout()  # 自动调整子图参数，以使所有子图适应图形区域
plt.show()
```

```python
# 绘制多个折线图
plt.figure(figsize=(10, 6))
subjects = ['语文', '数学', '英语']
for subject in subjects:
    # print(subject)
    plt.plot(df.index, df[subject], marker='o', label=subject) # 循环绘制，df.index为月份、subject为语文 数学 英语

plt.title('张三成绩月变化情况')
plt.xlabel('月份')
plt.ylabel('分数')
plt.legend()
plt.grid(True)

# 显示图形
plt.xticks(rotation=45) # 将 X 轴刻度标签的角度旋转为 45 度
plt.tight_layout()  # 自动调整子图参数，以使所有子图适应图形区域
plt.show()
```

```python
# 绘制多个柱状图
import pandas as pd
import matplotlib.pyplot as plt
# 读取 Excel 表格数据
excel_file = '张三成绩.xlsx'
df = pd.read_excel(excel_file, index_col=0)
# 绘制柱状图
plt.figure(figsize=(12, 6))
# 获取月份
months = df.index
# 设置每个月份的位置
x = range(len(months)) # x=range(12) 0-11
# 设置柱状图宽度
bar_width = 0.25
# 绘制张三在三门科目上的成绩柱状图 x=[0,1,2,3...]  \\\ i + bar_width for i in x = 0.25,1.25,2.25 \\\i + 2 * bar_width for i in x = 0.5,1.5,2.5...
plt.bar(x, df['语文'], width=bar_width, align='center', label='语文', color='skyblue') # align='center'  柱子对齐方式为居中
plt.bar([i + bar_width for i in x], df['数学'], width=bar_width, align='center', label='数学', color='r') #x 轴的位置，即月份，通过偏移量实现柱子的横向偏移
plt.bar([i + 2 * bar_width for i in x], df['英语'], width=bar_width, align='center', label='英语', color='lightgreen')
# 设置图形标题和坐标轴标签
plt.title('张三成绩月变化情况')
plt.xlabel('月份')
plt.ylabel('成绩')
# 设置 X 轴刻度标签
plt.xticks(x, months, rotation=45)  # 旋转45度，防止与月份重叠
# 添加图例
plt.legend()
# 显示图形
plt.tight_layout()
plt.show()
```

## 绘制散点图 scatter

```python
# 绘制散点图
import matplotlib.pyplot as plt # 绘制散点图
import numpy as np
# 生成样本数据
np.random.seed(21)
height = np.random.normal(loc=170, scale=10, size=100)  # 身高数据：正态分布，平均170cm，标准差10cm，生成100个样本
# height = np.random.randn(100) # 默认生成为均值为0.标准差为1

# 体重数据：与身高呈线性关系，斜率0.5，随机扰动为正态分布（平均0，标准差5kg）
weight = 0.5 * height - 20 + np.random.normal(loc=0, scale=5, size=100)
# 绘制散点图
plt.figure(figsize=(8, 6))
# height: 横坐标，weight: 纵坐标
# color: 散点颜色
# alpha: 透明度，0~1，值越小越透明
plt.scatter(height, weight, color='skyblue', alpha=0.9)
# 添加标题和坐标轴标签
plt.title('身高与体重关系散点图')
plt.xlabel('身高 (cm)')
plt.ylabel('体重 (kg)')
# 显示图形
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ----------------------
# 模拟训练数据
# ----------------------
epochs = np.arange(1, 101)  # 训练 100 个 epoch

# Loss 模拟：训练集和验证集损失随 epoch 下降
train_loss = np.exp(-0.05 * epochs) + 0.05 * np.random.rand(100)  # 指数下降 + 小扰动
val_loss = np.exp(-0.045 * epochs) + 0.05 * np.random.rand(100)   # 略慢下降

# Accuracy 模拟：训练集和验证集准确率随 epoch 上升
train_acc = 1 - np.exp(-0.05 * epochs) + 0.02 * np.random.rand(100)
val_acc = 1 - np.exp(-0.045 * epochs) + 0.02 * np.random.rand(100)

# ----------------------
# 创建画布
# ----------------------
plt.figure(figsize=(14, 6))  # 宽 14，高 6

# ----------------------
# 绘制 Loss 曲线
# ----------------------
plt.subplot(1, 2, 1)  # 1行2列，第1个子图
plt.plot(epochs, train_loss, 'r-', label='Train Loss')
plt.plot(epochs, val_loss, 'b--', label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss (100 Epochs)')
plt.legend()
plt.grid(True)

# ----------------------
# 绘制 Accuracy 曲线
# ----------------------
plt.subplot(1, 2, 2)  # 1行2列，第2个子图
plt.plot(epochs, train_acc, 'g-', label='Train Accuracy')
plt.plot(epochs, val_acc, 'm--', label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy (100 Epochs)')
plt.legend()
plt.grid(True)

# 自动调整子图间距
plt.tight_layout()

# 显示图形
plt.show()
```
