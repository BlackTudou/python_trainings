```python
%matplotlib inline 
#若不显示图片 则加上
from matplotlib import pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

| 参数       | 全称          | 作用       | 常见取值                                            |
| -------- | ----------- | -------- | ----------------------------------------------- |
| `ls`     | `linestyle` | 线型       | `'-'` 实线、`':'` 点虚线、`'--'` 破折线、`'-.'` 点划线        |
| `lw`     | `linewidth` | 线宽       | 数值，如 `1.5`                                      |
| `label`  | 标签          | 图例显示的文字  | `"sin(x)"`                                      |
| `color`  | 颜色          | 设置线条颜色   | `'r'` 红、`'g'` 绿、`'b'` 蓝、`'y'` 黄、`'k'` 黑、`'w'` 白 |
| `marker` | 标记点         | 每个数据点的样式 | `"."` 点、`","` 像素点、`"o"` 圆点、`"v"` 下三角、`"*"` 星号   |


```python
# 导入 Matplotlib 的 pyplot 模块，并简写为 plt
# pyplot 提供了常用的绘图函数
import matplotlib.pyplot as plt  
# from matplotlib import pyplot as plt
# 定义横坐标数据
x = [1, 2, 3, 4, 5]  

# 定义纵坐标数据
y = [2, 4, 6, 8, 10]  

# 绘制折线图（直线图）

plt.plot(x, y, marker="*",color = "b",ls = "-",label = "y=2x")
# 参数说明：
# x, y       → 数据
# ls=":"     → 线型为点虚线（linestyle）
# color="r"  → 红色线条（red）
# marker="*" → 每个数据点显示为星型标记
# label="Huitu" → 图例名称
# plt.plot(x, y, ls=":", color="r", marker="*", label="Huitu")  

# plt.plot(x, y, color="r", ls="--")   # 红色虚线
# plt.plot(x, y, 'r--')                # 简写形式：红色虚线

# 显示图例（legend），会根据 label 参数自动生成
plt.legend()  

# 显示绘制好的图形（在 Jupyter Notebook 中可省略）
plt.show()
```
