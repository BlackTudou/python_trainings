### python中print的语法结构

```python
print(1,2,3,4,5,6) # sep = 空格
print(1,2,3,4,5,6, sep="->") # sep = "->"
```

```python
print(1,2,3,4,5,6) # end 默认为\n
print(1,2,3,4,5,6)

print(1,2,3,4,5,6, end="\t") # end 改为"\t"(TAB 键)
print(1,2,3,4,5,6)
print(1,2,3,4,5,6)
```

```python
f = open("test.txt", "a") # 打开文件，如果文件不存在，则创建文件，a 追加内容, w 覆盖写入
print("hello world", file=f)
f.close()
```

### print 三种格式化输出

```python
print("name: %s, age: %d" % ("张三", 20))
```

```python
print("name:{}, age:{}".format("张三", 20))
print("name:{0}, age:{1}".format("张三", 20))
print("name:{1}, age:{0}".format("张三", 20))
print("name:{name}, age:{age}".format(name="张三", age=20))
print("name:{name}, age:{age}".format(age=20, name="张三"))
```

```python
"""
f表达式输出，最常用
"""
print(f"name:{'张三'}, age:{20}")

name = "张三"
age = 20
print(f"name:{name}, age:{age}")

name = "李四"
age = 22
print(f"name:{name}, age:{age}")
```

### 变量与对象

```python
"""
python对象三要素
1.id 内存地址
2.type 类型
3.value 值
"""

a = 1 # 1 是int类型对象，a是变量名，a指向1这个对象
print(type(a)) # 查看a的类型
print(id(a)) # 查看a的内存地址

a = 2 # 2 是int类型对象，a是变量名，a指向2这个对象
print(id(a)) # 查看a的内存地址，a的内存地址发生了变化，说明a指向了新的对象2
```

```python
a = 10
b = a
print(id(a)) # 2346797656656
print(id(b)) # 2346797656656
```

```python
epoch = 100
acc = 1.0
print(type(epoch))
print(type(acc))
print(f"epoch:{epoch}, acc:{acc}")
```

### input输入函数

```python
name = input("请输入你的名字：")
print(name)
print(type(name)) # input返回的都是字符串类型
```

```python
age = int(input("请输入你的年龄：")) # 类型的强制转换，将输入的字符串转换为整型
print(age)
print(type(age)) # 经过强制转换后，age的类型为int
```
