### ndarray对象

ndarray 是一个n维数组对象，创建 ``ndarray`` 对象只需要调用 ``numpy.array()`` 即可

```python
# 创建 ndarray 对象

import numpy as np
a = np.array([1,2,3]) # <class 'numpy.ndarray'>
# a = np.array((1,2,3)) # 两者均可
print(a)
print(type(a))
```
