## 定义张量

```python
# 张量的创建
import torch
a = torch.Tensor([[1, 2, 3], [4, 5, 6]])
print(type(a))#<class 'torch.Tensor'>
print(a.shape)
a
```

```python
a = torch.eye(2) # 对角阵
a
```

```python
b = torch.zeros(3,3)
b
```

```python
c = torch.ones(3,3)
c
```

```python
d = torch.arange(1,10,1) # start end step 左闭右开
d
```

```python
e = torch.linspace(1,10,10) # 使用等差函数创建张量
e
```

```python
# 等比数列
f = torch.logspace(1,10,10) # 基数为10 10^1-10^10
f
```

```python
g = torch.logspace(1,2,10) # 10^1-10^2  num=10
g
```

## 张量的索引

```python
import torch
a = torch.Tensor(2,3,128,128) # 2张图片，每张图片3通道(RGB)，高宽为128x128
print(a.shape)
print(a[0].shape) # 取第一张图像 torch.Size([3, 128, 128])
print(a[0][0].shape) # 取第一张图像的第一通道 torch.Size([128, 128])
print(a[0][0][0].shape) # torch.Size([128])

a1 = torch.rand(64, 10) # 随机创建张量
print(a1.shape)
print(a1[63].shape)
```

## 张量的切片

```python
#tensor[start:end:step]

import torch
a = torch.Tensor(2,3,128,128) # 2张图片，每张图片3通道(RGB)，高宽为128x128
print(a.shape)
print(a[:,:,:,:].shape)
print(a[:1,:,:,:].shape) # 取第一张图
print(a[:1,:1,:,:].shape) # 取第一张图的第一个通道
print(a[:1,:1,:64,:64].shape) # torch.Size([1, 1, 64, 64]) 取第一张图的第一个通道的左上四分之一
print(a[:1,:1,:64:2,:64:2].shape) # torch.Size([1, 1, 32, 32]) 在上面基础上跳着一个像素点取
```

```python
a = torch.randn(2,3,128,128)
print(a[:1,:1,:1,:1])  # 取第一张图的第一个通道 第一个像素点 像素值
```

## 维度变换

```python
# 使用reshape改变张量
a = torch.randn(2,3,128,128) # 使用reshape改变张量
b = a.reshape(2,3,-1) # 4维->3维    16384=128x128
print(b.shape)
c = a.reshape(4,-1) # 4维->2维
print(c.shape) # 24576=(2*3*128*128)/4
```

```python
# 维度的增加
a = torch.randn(2,3,128,128) # 维度的增加
a = a.unsqueeze(2) # a.unsqueeze(0).shape  维度添到维度2的前面（维度0,1,2...）
a.shape
```

```python
a = torch.randn(2,3,128,128) # 维度的删减 只能删除大小为一的维度
a = a.squeeze(1)
print(a.shape)

b = torch.randn(2,1,128,128) # 维度的删减 只能删除大小为一的维度
b = b.squeeze(1)
print(b.shape)
```

## 维度调整

```python
a = torch.randn(2,3,128,128) # 交换维度，且只能交换张量中某两个维度的顺序
print(a.shape)
a = a.transpose(0,2)
print(a.shape)
```

```python
a = torch.randn(2,3,128,64) # torch.Size([3, 2, 64, 128]) 多个维度的交换
print(a.shape)
a = a.permute(1,0,3,2)
print(a.shape)
```

```python
a = torch.randn(2,1,128,64) # 将一个维度大小为1的维度扩展为其他大小
print(a.shape)
a = a.expand(2,3,128,64)
print(a.shape)
```
