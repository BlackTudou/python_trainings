### 选择（分支）结构

```python
# if 单分支结构
a = 10
if a > 9:
    print("success")
```

```python
# if 双分支结构
a = -1
if a >= 0:
    print(f"a的值为非负，a的值为{a}")
else:
    print(f"a的值为负，a的值为{a}")
```

```python
# if 多分支结构
# and or not
score = int(input("请输入成绩："))

if score < 0 or score > 100:
    print("输入成绩有问题，请重新输出")
elif score >=90 and score <= 100:
    print("该学生成绩为A")
elif score >= 80 and score < 90:
    print("该学生成绩为B")
elif score >= 70 and score < 80:
    print("该学生成绩为C")
elif score >= 60 and score < 70:
    print("该学生成绩为D")
else:
    print("该学生成绩不及格")
```

```python
# if else 的嵌套

age = 18
money = 8

if age >= 18:
    print("年龄已经达到18岁，可以进入网吧啦~~")
    if money >=10:
        print("身上的钱够上网了，可以愉快的打游戏了")
    else:
        print("虽然我满18岁了，但是没有钱上不了网")
else:
    print("年龄未满18岁，回家洗洗睡吧")
```

### 循环结构

```python
# while 循环

a = 0

while a < 10:
    print(a)
    a += 1
```

```python
# 使用while循环实现一个九九乘法表

"""
1*1=1
1*2=2 2*2=4
1*3=3 2*3=6 3*3=9

横着循环
"""

i = 1
while i < 10:
    j = 1
    while j < i + 1:
        print(f"{j}*{i}={j * i}", end="\t")
        j += 1
    print(end="\n")
    i += 1
```

```python
# for 循环
# for 变量名 in 可迭代对象:

for i in range(10): # range(10) 0-9
    print(i)
```

```python
# 使用for循环实现一个九九乘法表

for i in range(1, 10):
    for j in range(1, i + 1):
        print(f"{j}*{i}={j*i}", end="\t")
    print()
```

### 函数的定义和调用

```python
# 定义一个相加的函数

def add(a, b):
    return (a+b)

sum = add(3, 4)
print(sum)
```

```python
# 匿名函数
# 函数名 关键词(lambda) 参数 表达式
add = lambda a,b:a+b
add(4, 5)
```

### 局部变量和全局变量

```python
# 局部变量和全局变量

a = 10 # 全局变量
def test():
    a = 2
    print(a)

test()
print(a)
```

### 函数的递归

```python
# 使用递归实现N的阶乘
# e.g. 5!=5*4*3*2*1

def factorial(n):
    if n == 1:
        return 1
    else:
        return n*factorial(n-1)

factorial(5)
```
