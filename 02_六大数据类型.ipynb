## 数值类型

### 整形

```python
a = 1
print(a)
print(type(a))
```

```python
# 加减乘除
a = 1
b = 2
c = a + b
d = a - b
e = a * b
f = a / b

print(c, d, e, f) # 加减乘除运算
print(type(c), type(d), type(e), type(f)) # 加减乘除运算结果的类型
```

```python
# 取余运算
a = 11
b = 3
c = a % b # 取余运算

print(c)
print(type(c))
```

```python
# 整除运算
a = 10
b = 3
c = a / b # 除法运算 会保留小数
d = a // b # 整除 只保留整数
print(c, d)
```

```python
# 求幂运算
a = 2
b = 3
c = a ** b # 求幂运算 2的3次方
print(c)
```

### 浮点型

```python
a = 1.11
b = 1.12
c = a * b
print(type(c))
```

```python
# 浮点类型不精确存储
# 浮点类型格式化输出，保留制定小数位
a = 0.1
b = 0.2
c = a + b
print(c)
print("a+b的值为：%f" % (c)) # %f默认保留六位小数
print("a+b的值为：%.3f" % (c)) # %.3f保留三位小数
```

```python
# 科学计数法
a = 10e3 # 10x10^3
b = 1e3 # 1x10^3
c = 1e-3 # 1x10^(-3)=0.001
print(type(a), type(b), type(c))
print(a, b, c)
```

### bool 类型

```python
print(True)
print(False)
print(True + 1) # 为2 代表True=1
print(False + 1) # 为1 代表False=0
print(type(True))
```

```python
"""
True：非空、非零数值、True
False：None、0、空字符串、False
"""
print(bool(""))
print(bool(None))
```

### 复数类型

```python
a = 1 + 2j
print(a)
print(type(a)) # <class 'complex'>

b = complex(1, 2) # 复数的函数定义
print(b)
```

## 字符串类型

```python
# 字符串定义，单引号，双引号，三引号均可
str1 = "hello world"
print(str1)
print(type(str1))

str2 = 'hello world'
print(str2)
print(type(str2))

str3 = """hello world"""
print(str3)
print(type(str3))
```

```python
# 字符串索引查询
str = "hello world"
print(f"str[6]={str[6]}") # 索引查询，空格也有索引
print(f"str[-3]={str[-3]}") # 索引查询 从右往左 -1,-2,-3
```

```python
# 字符串切片
str = "hello world"

str1 = str[0:-1:1] #[start:end:step] 左闭右开
print(str1) # hello worl 没取到d

str2 = str[0::1]
print(str2) # hello world

str3 = str[:] # 或者 str[::]
print(str3) # hello world

str4 = str[1:-1:1] # 截取 ello worl
print(str4) # ello worl

str5 = str[1:-1:2] # 截取 ello worl，step=2
print(str5) # el ol
```

```python
# 字符串是不可变数据类型
str = "hello world"
print(id(str))
# str[0] = "a" # 错误
str = "aello world" # 对str重新赋值
print(str)
print(id(str))
```

```python
# 字符串相加
a = "hello "
b = "word"
c = a + b
print(c)
```

## 列表类型

```python
# 列表定义
lst = [1, 2, "hello", [1, 2, 3]]
print(lst)
print(type(lst))
```

```python
# 列表查询
lst = [1, 2, "hello", [1, 2, 3]]
print(lst[-1], lst[0]) # 查询
```

```python
# 列表切片
lst = [1, 2, 3, "hello", [1, 2, 3]]

lst1 = lst[:] # 或者 lst[::] 默认lst[0:-1:1]
lst2 = lst[3:] # 取 "hello", [1, 2, 3]   相当于lst[3:-1:1]
lst3 = lst[-4:-1] # 取 2, 3, "hello"，从负索引开取， 左闭右开

print(lst1)
print(lst2)
print(lst3)
```

```python
# 列表嵌套
lst = [1, 2, 3, "hello", [1, 2, 3]]
lst1 = lst[-1]
print(lst1)
print(type(lst1))

lst2 = [[1, 2],
        [3, 4]]
print(lst2)
```

```python
# 列表增加
lst = [1, 2, 3, "hello", [1, 2, 3]]
lst.append("world")
print(lst)
```

```python
# 列表删除
lst = [1, 2, 3, "hello", [1, 2, 3]]
lst.remove("hello")
print(lst)
```

```python
# 列表修改元素，直接用索引改
lst = [1, 2, 3, "hello", [1, 2, 3]]
lst[2]= "nihao"
print(lst)
```

```python
# list 函数
lst = list(range(10)) # range(10) [0:10:1]
print(lst)
```

## 元组类型 Tuple

```python
# 元组定义
a = (1, 2, 3)
print(a)
print(type(a))

b = (1) # 单一用括号并不是元组类型
c = (1,) # 单一元素需要加上逗号，才是元组
print(b, c)
print(type(b), type(c))
```

```python
# 元组查询 切片
# 元组跟字符串一样，为不可更改数据类型，只可查看
a = (1, 2, 3)
b = a[0]
c = a[:]
d = a[1:] # 元组切片

print(a)
print(b)
print(c)
print(d)
```

```python
# 元组函数
a = (1, 1, 2, 1, 3, 2)
print(a.count(1)) # 统计元组a中1的count
print(a.index(2)) # 2最先出现的index
print(f"max(a)={max(a)}") #元组a中的最大值
print(f"len(a)={len(a)}") #元组a的长度
```

## 集合 Set

```python
# Set 是一个不重复元素的序列，里面存储的元素是唯一的
# 集合不能查，改，因为无序的，不支持下标操作

# 集合的定义
a = {1, 2, 3, 4, 5, "hello", 1, 22}
print(a) # 重复的1就没了，集合会自动去重
print(type(a))
```

```python
# 空集合的定义
a = {}
print(a)
print(type(a)) # 空的{}代表字典，不是集合

b = set() # 空集合要用set函数定义
print(type(b))
```

```python
# 集合增加元素
a = {1, 2, 3, "hello", 1, 22}
a.add(6)
print(a)

# 集合删除元素
a.remove(2)
print(a)
```

```python
# 集合交集与并集
a = {1,2,3,4}
b = {3,4,5,6}
c = a & b # 交集
d = a | b # 并集

print(c)
print(d)
```

## 字典 Dictionary

```python
# 字典定义
dict = {"name":"zhangsan", "age":20, "height":180.2}
print(dict)
print(type(dict))
```

```python
# 增加元素
dict = {"name":"zhangsan", "age":20, "height":180.2}
dict["weight"] = 150
print(dict)
```

```python
# 删除元素
dict = {"name":"zhangsan", "age":20, "height":180.2, "weight":150}
del dict["age"]
print(dict)
```

```python
# 修改元素
dict = {"name":"zhangsan", "age":20, "height":180.2, "weight":150}
dict["age"] = 18
print(dict)
```

```python
# 查询元素
dict = {"name":"zhangsan", "age":20, "height":180.2, "weight":150}
age = dict["age"]
print(age)
```
